El algoritmo de optimización de enjambre, también conocido como algoritmo de optimización de colonias de hormigas, es un método de búsqueda inspirado en el comportamiento colectivo de los insectos sociales, como las hormigas. A continuación, te presento una breve lista que describe las características principales de este algoritmo:

   1. Inicialización: Se crea un conjunto de soluciones iniciales aleatorias, también conocidas como "hormigas virtuales".

   2. Evaluación: Cada solución es evaluada utilizando una función objetivo que determina su calidad en relación con el problema que se está abordando.

   3. Actualización local: Las soluciones individuales ajustan su comportamiento según las mejores soluciones locales encontradas por las hormigas cercanas.

   4. Movimiento: Las soluciones se desplazan en el espacio de búsqueda utilizando una regla de transición basada en probabilidades. Las hormigas virtuales siguen rastros de feromonas, que se depositan y evaporan en el entorno.

   5. Actualización global: Se actualizan los rastros de feromonas de acuerdo con la calidad de las soluciones encontradas. Las mejores soluciones tienen una influencia mayor en la actualización de los rastros.

   6. Iteración: Los pasos 3 a 5 se repiten hasta que se cumple un criterio de terminación, como un número máximo de iteraciones o la convergencia de las soluciones.

   7.  Explotación y exploración: El algoritmo combina la explotación de las mejores soluciones conocidas con la exploración de nuevas regiones del espacio de búsqueda, permitiendo la búsqueda de soluciones óptimas globales.

   8. Convergencia: El algoritmo converge hacia una o varias soluciones óptimas a medida que las hormigas virtuales intercambian información y encuentran mejores soluciones a lo largo de las iteraciones.

```python
import numpy as np
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
```

```python
# Definimos la función objetivo a optimizar
def f(x):
    return np.sum(x**2, axis=1)

# Configuramos el optimizador
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}

# Creamos una instancia del optimizador
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)

# Ejecutamos el optimizador para minimizar la función
cost, pos = optimizer.optimize(f, iters=1000, verbose=2)
```

```python
print(f'El mínimo de la función se encuentra en: {pos}, y el costo es: {cost}')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Creamos una grilla de puntos en el espacio 2D
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

# Calculamos el valor de la función en cada punto
z = x**2 + y**2

# Creamos una figura 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficamos la función
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.2)

# Graficamos el mínimo encontrado
ax.scatter(pos[0], pos[1], cost, color='r', s=100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
